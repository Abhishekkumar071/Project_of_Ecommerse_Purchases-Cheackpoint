```python
import pandas as pd  # For data manipulation
import numpy as np   # For numerical operations
import matplotlib.pyplot as plt  # For plotting graphs
import seaborn as sns  # For enhanced data visualization
```

```python
orders = pd.read_csv('C:\\Users\\abhis\\Videos\\Orders (1).csv')  # Load orders data from CSV file
customers = pd.read_csv('C:\\Users\\abhis\\Videos\\Customers.csv')  # Load customers data from CSV file
restaurants = pd.read_csv('C:\\Users\\abhis\\Videos\\Restaurants (1).csv')  # Load restaurants data from CSV file
```

```python
orders  # Display the contents of the orders DataFrame
```

```python
orders.info()  # Display a summary of the orders DataFrame, including data types and non-null counts
```

```python
orders["Order_Date"] = pd.to_datetime(orders["Order_Date"])  # Convert the Order_Date column to datetime format
```

```python
orders.info()
```

```python
orders["Month"] = orders["Order_Date"].dt.strftime("%B")  # Extract the full month name from Order_Date
orders  # Display the updated DataFrame with the new Month column
```

```python
plt.figure(figsize=(6, 3))  # Set the figure size to 6 inches wide and 3 inches tall
sns.countplot(x="Month", data=orders, color="blue")  # Create a count plot for the Month column
plt.xticks(rotation=30)  # Rotate x-axis labels by 30 degrees for better readability
plt.show()  # Display the plot
```

```python
orders["Hour"] = orders["Order_Date"].dt.strftime("%H")  # Extract the hour from Order_Date
orders["Hour"] = orders["Hour"].astype("int")  # Convert the Hour column to integer type
orders = orders.sort_values(by="Hour")  # Sort the DataFrame by Hour
sns.lineplot(x="Hour", y="Total_Amount", data=orders, estimator="sum")  # Create a line plot
plt.show()  # Display the plot
```

```python
# Merge the 'orders' DataFrame with the 'customers' DataFrame
# on the 'Customer_ID' column using an inner join
df = pd.merge(left=orders, right=customers, on="Customer_ID", how="inner")

# Display the merged DataFrame
df
```

```python
# Set the figure size for the plot (width, height)
plt.figure(figsize=(12, 5))

# Create a count plot to visualize the number of customers by location, 
# with different colors for each age group
sns.countplot(x="Customer_Location", data=df, hue="Customer_Age_Group", palette="viridis", edgecolor="black")

# Adjust the legend position to the upper right corner of the plot
plt.legend(bbox_to_anchor=(1, 1))

# Display the plot
plt.show()
```

```python
#This visualization will show how ratings are distributed across all restaurants.

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Create a histogram for Ratings
plt.figure(figsize=(10, 6))
sns.histplot(restaurants['Ratings'], bins=10, kde=True, color='blue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sort the restaurants by Ratings and get the top 10
top_restaurants = restaurants.sort_values(by='Ratings', ascending=False).head(10)

# Create a bar plot for top restaurants by ratings without the palette
plt.figure(figsize=(12, 6))
sns.barplot(x='Ratings', y='Name', data=top_restaurants)  # Removed palette argument
plt.title('Top 10 Restaurants by Ratings')
plt.xlabel('Ratings')
plt.ylabel('Restaurant Name')
plt.show()
```
